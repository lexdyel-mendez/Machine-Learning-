## Importing necessary libraries 


```python
import math
import matplotlib.pyplot as plt
```

## Mean Value 

Gives the averege value of the set. The formula used for this function is:


```python
def mean(sampleset):
    total = 0
    for element in sampleset:
        total = total+element
    return total/len(sampleset)
```

## Variance 

Explains how dispereced values are on the dataset. The grater the number is, the more scattered the values are. The same needs the mean value as a starting point. The formula for the variance is:


```python
def variance(sampleset):
    total = 0
    setmean = mean(sampleset)
    for element in sampleset:
        total = total + (math.pow(element-setmean, 2))
    return total/len(sampleset)
```

## Testers 

This section is saved for testing each method.

## Mean value tester

```python
myset = [2, 4, 7, 3, 9 ,1, 4]
myset2 = [1, 10, 20, 4, 75, -100, 30]
mymean = mean(myset)
mymean2 = mean(myset2)
```

```python
plt.figure(figsize=(16,8))
plt.plot(myset)
plt.plot([mymean] * len(myset))
```

```python
plt.figure(figsize=(15,8))
plt.plot(myset2)
plt.plot([mymean2]*len(myset2))
```

## Variance value tester

```python
print('Variance of the first data set:',variance(myset),'\n')
print('Variance of the second data set:',variance(myset2))
```
