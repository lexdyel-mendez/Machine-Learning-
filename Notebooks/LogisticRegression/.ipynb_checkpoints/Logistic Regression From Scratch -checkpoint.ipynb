# Create the data Set

Creates 2 chunks of features that are divided with a lable of either a 1 or a zero

```python
np.random.seed(12)
num_observations = 5000

x1 = np.random.multivariate_normal([0, 0], [[1, .75],[.75, 1]], num_observations)
x2 = np.random.multivariate_normal([1, 4], [[1, .75],[.75, 1]], num_observations)

simulated_separableish_features = np.vstack((x1, x2)).astype(np.float32)
simulated_labels = np.hstack((np.zeros(num_observations),
                              np.ones(num_observations)))

Data_with_intercepts = np.hstack((np.ones((simulated_separableish_features.shape[0],1)),
                    simulated_separableish_features))
```

### Create the Sigmoid function

Create the sigmoid function that is used on the hypothesis

```python
def sigmoid(theta):
    return 1 / (1 + np.exp(-theta))
```

### Visualization of the Sigmoid Function

```python
x = np.arange(-10, 10, .2)
sig = sigmoid(x)

ax = plt.figure(figsize=(14, 8)).add_subplot(1, 1, 1)
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.plot(x, sig)
```

### Calculating the Log-likelihood

The likelihood function is always defined as a function of the parameter  {\displaystyle \theta } \theta equal to (or sometimes proportional to) the density of the observed data with respect to a common or reference measure, for both discrete and continuous probability distributions. 

```python
def log_likelihood(features, target, weights):
    scores = np.dot(features, weights)
    ll = np.sum( target*scores - np.log(1 + np.exp(scores)) )
    return ll
```

### Create the Hypothesis

Create the hypothesis function that uses the sigmoid

```python
def hypothesis(features, theta):
    z = np.dot(features, theta)
    return sigmoid(z)
```

## Creating the Gradient Descent 

Create the method that controls the gradient descent

```python
def GradientDescent(features, output):
    return np.dot(features.T, output)
```

### Creating the Logistic Regression model

```python
def logistic_regression(features, target, num_iter, learning_rate, add_intercept = False):
    if add_intercept:
        intercept = np.ones((features.shape[0], 1))
        features = np.hstack((intercept, features))
        
    theta = np.zeros(features.shape[1])
    
    for step in range(num_iter):
        predictions = hypothesis(features, theta)

        # Update weights with gradient
        output_error_signal = target - predictions
        theta += learning_rate * GradientDescent(features, output_error_signal)
        
        # Print log-likelihood every so often
        if step % 1000== 0:
            print ("Log likelihood at step" , step, "is",log_likelihood(features, target, theta))
        
    return theta
```

### Testing the model

```python
theta = logistic_regression(simulated_separableish_features, simulated_labels,
                     num_iter = 50000, learning_rate = .001, add_intercept=True)
```

# Comparison with Sklearn

Compare it with the Sklearn Logistic Regression. 

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(fit_intercept=True, C = 1e15)
clf.fit(simulated_separableish_features, simulated_labels)

print ("Weigths of the Sklearn model:",clf.intercept_, clf.coef_)
print ("Weigths of the scratch model:",theta)
```

## Accuracy

Accuracy of both types of Logistic Regression

```python
final_scores = np.dot(Data_with_intercepts, theta)
preds = np.round(sigmoid(final_scores))

print ('Accuracy from scratch: {0}'.format((preds == simulated_labels).sum().astype(float) / len(preds)*100))
print ('Accuracy from sk-learn: {0}'.format(clf.score(simulated_separableish_features, simulated_labels)*100))
```
